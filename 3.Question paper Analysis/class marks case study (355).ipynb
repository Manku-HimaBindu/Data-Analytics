# Dataset Description:
The dataset contains 86 student records with 12 columns representing total marks and marks scored in different questions. 

The columns are:

Total - The total marks scored by a student.

Q1aM4 - Marks for Question 1a (out of 4).

Q1bM6 - Marks for Question 1b .

Q2aM6 - Marks for Question 2a .

Q2bM4 - Marks for Question 2b .

Q3aM5 - Marks for Question 3a .

Q3bM5 - Marks for Question 3b .

Q4aM3 - Marks for Question 4a .

Q4bM7 - Marks for Question 4b .

Q5M10 - Marks for Question 5 .

Q6aM4 - Marks for Question 6a .

Q6bM6 - Marks for Question 6b .

```python
df["Percentage"]=df.Total/60 *100
df.fillna(0,inplace=True)
```

```python
def assign_grade(percentage):
    if percentage >= 61:
        return 'O'
    elif percentage >= 55:
        return 'A+'
    elif percentage >= 45:
        return 'A'
    elif percentage >= 30:
        return 'B'
    else:
        return 'F'

df['Grade'] = df['Percentage'].apply(assign_grade)

df
```

```python
grade_counts = df["Grade"].value_counts()
plt.figure(figsize=(8, 5))
grade_counts.plot(kind="bar", color="skyblue")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.xticks(rotation=0)
plt.show()
```
